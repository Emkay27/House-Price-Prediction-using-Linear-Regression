## Overview

Regression analysis is a fundamental concept in the field of machine learning. It falls under supervised learning wherein the algorithm is trained with both input features and output labels. It helps in establishing a relationship among the variables by estimating how one variable affects the other.

#### Linear Regression:

Linear regression finds the linear relationship between the dependent variable and one or more independent variables using a best-fit straight line. Generally, a linear model makes a prediction by simply computing a weighted sum of the input features, plus a constant called the bias term (also called the intercept term). In this technique, the dependent variable is continuous, the independent variable(s) can be continuous or discrete, and the nature of the regression line is linear.

## Approach

We use scikit-learn to implement Linear regression on the housing price dataset to predict the price of the houses in the city based on the income of the residents, age of the houses, number of rooms in the house, number of bedrooms in the house, address and area population in the city.

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
%matplotlib inline
```

## Dataset

The dataset used here is a house price dataset found from Kaggle, because house price is a continuous variable, then this is a regression problem. The dataset consists of the address of a house, number of houses in the city, Price that a house is sold for, number of rooms in the house, age of the houses in the same city and the income of the residents in the city.

#### Correlation Matrix Plot

Since the dataset has many columns, we use the correlation matrix as a heatmap to quickly get a visualization of the correlations among columns, on the heatmap below, we see that green means positive and red means negative, so the stronger the color the larger the correlation magnitude. The main diagonal shows the strongest correlated pairs, we can see that Avg_Area_Income, Avg_Area_House_Age and Area_population are respectively the three most correlated variables with the Price.

```python
plt.figure(figsize=(10,10))
sns.heatmap(house_pricing.corr(), annot = True, cmap = 'RdYlGn', linewidth = 1)
plt.show()
```

#### Describing the dependent and Independent variables

Here X_dep represents the dependent variables, Y_indep represents the independent variable.

```python
X_dep = house_pricing[['Avg_Area_Income', 'Avg_Area_House_Age', 'Avg_Area_Number_of_Rooms',
                      'Area_Population']]
Y_indep = house_pricing['Price']
```

#### Creating a train and test dataset


We split the dataset into training and testing sets respectively and then train with the training set, in doing so we can have an accurate evaluation of the out-of-sample accuracy because the testing set is not part of the dataset that we used to train the model. Here 80% of the dataset is for training and the remaining 20% is for testing, we use 42 for the random state.

```python
x_train, x_test, y_train, y_test = train_test_split(X_dep, Y_indep, test_size = 0.2, random_state = 42)
```
